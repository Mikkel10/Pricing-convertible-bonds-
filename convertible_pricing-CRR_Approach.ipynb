# Abstract

This article tries to explore the relationship between the theoretical binomial-tree price and the real market price of all convertible bonds issued in China. The goal is to evaluate whether the binomial-tree price can accurately measure the market price.

The main conclusions are as followed:

(1)	On the first day of listing, the binomial-tree price and the real market price maintain the stationary relationship. But binomial-tree price systematically undervalued the real price for about 10%.

(2)	In the duration of a single convertible bond, on average, the binomial-tree price and the market price are non-stationary: in the early stage of listing, binomial-tree price is below the market price; with time going on, binomial-tree price gradually becomes closer to the market price.

(3)	By adopting a strategy that takes advantage of the maximum deviation between the binomial-tree price and the market price (if we assume them to become equal eventually), we could achieve 9% annualized return historically. But according to the results of back-testing, the historical returns are highly volatile, indicating that the strategy is not ideal.

(4)	Since the facts that the number of listed convertible bonds in China is limited (around 300 bonds in total) and that they massively went into open market during 2017-2019 and were strongly influenced by the bull market during this period, the relationship between the binomial-tree price and the market price is unstable as we expected. Although in general it is hard to measure the market price with binomial-tree price, this model still has some predicting power for some individual bond.


# Introduction

Convertible bonds are hybrid securities issued by a firm where the holder has the right to convert the bond into the common stocks of the firm according to pre-specified conditions.

The market of convertible bonds is booming in China. For the first half of 2019, there are 62 convertible bonds, worth of 20 billion USD, that went into public. The number of bonds increased by 210% and the amount of money increased by 570% over the same period last year. The latter even exceeded last year's total amount. Accordingly, there are tons of capital pouring into this market. Some have claimed that the new era of the market of convertible bonds has come in China.

There are many reasons behind this boom. Firstly, since Chinese government put more restrictions on financing methods during 2018, listed company have to resort to other ways, such as convertible bonds, to fund themselves. In addition, the embedded option to convert bonds to stocks makes convertible bonds more stock-like and easier to speculate. Compared to those of the Chinese A-share stock market, the characters of the convertible bonds market, such as T+0 trading system and no limit on the rise of fall in a single trading day, are more favorable for many institutional investors. 

Thanks to the pioneer works in financial engineering, such as Black-Scholes-Merton Option Pricing Model, it is no longer a difficult task to price the intrinsic value of simplified convertible bonds. Even with various provisions embedded into the convertible bonds, their values should be easier to accurately modeled than stocks. Nevertheless, market of convertible bonds in China seems to be pretty confused recently: dramatic change in prices and overreaction happen all the time. Most of the large financial investors trade convertible bonds based on fundamental analysis, rather than more specific pricing tools. 

Can we accurately price the convertible bonds listed in China? That’s the main question that the author want to address in the series of articles.


# Method Illustration

In theory, the value of typical listed convertible bonds in China = 
+ the value of bond 
+ +the value of conversion provisions (call option on the equity) 
- -the value of redemption provisions (call option on the equity) 
+ +the value of reset provisions (the conversion price is adjusted downwards if the underlying stock price does not exceed pre-specified prices) 
- -put provisions of the bond.

All the embedded options above are American options.

Given that the boundary of reset provisions and put provisions of the bond are hard to judge with binomial model and that they are not the main factors, this article only considers the first three parts. Further with Monte Carlo simulation, we could address the values of reset clause.

This article describes the relationship with the ratio (cal_to_real), which is the calculated price divided by the market price. If cal_to_real is above 1, the market price is undervalued; if cal_to_real is below 1, the market price is overvalued. 

Also, in the first sixth month after listing, the right to convert bonds to stocks could not be exercised. Hence when calculating bonds within this period, we must offset the value of conversion right at this time. 
For example, for the convertible bonds with 6-year duration, its value the first day of listing = 
+ bond value 
+ +the value of conversion right from present to the sixth year in the future  
- -the value of conversion right from present to the sixth month in the future 
- -the value of redemption right from present to the sixth year in the future 
+ +the value of redemption right from present to the sixth month in the future

The bond value is calculated via discounted cash flow method. The embedded American options are calculated via Cox-Ross-Rubinstein model(hereinafter referred to as CRR model).

We take the treasury yield at different time point as the discount rate. To mitigate the market influence on the volatility of certain point, the volatility of stock price, which is needed in CRR model, are assumed as annualized 65-day exponential average of the daily volatility. 

All the date comes from RiceQuant, an online quantitative trading platform(https://www.ricequant.com/welcome/#research).

```python
import rqdatac as rq
rq.init()
from rqdatac import *
import pandas as pd
import sys
import numpy as np
from utility import zwdata as zw
import QuantLib as ql 
import talib as ta
import math
import matplotlib.pyplot as plt
from multiprocessing import Pool
```

```python
def get_convertible_call_price(bond_name,date):
    '''
    Calculate the values of embedded options
    Assuming that the dividend yield equals to 0
    Assuming volatility as 65-day EMA of daily volatility
    The outputs are option values per ￥100 convertible bonds
    '''
    
    # Assuming that the dividend yield equals to 0
    dividend_rate = 0
    
    # Acquire each bond information
    all_info = rq.convertible.all_instruments()
    info = all_info[all_info['symbol'] == bond_name]
    order_book_id = info['order_book_id'].iloc[0]
    stock_code = info['stock_code'].iloc[0]
    maturity_date = info['maturity_date'].iloc[0]
    conversion_start_date = info['conversion_start_date'].iloc[0]
    value_date = info['value_date'].iloc[0]
    de_listed_date = info['de_listed_date'].iloc[0]
    
    # Acquire redemption multiple
    redumption = pd.read_excel('赎回条款.xlsx')
    redumption = redumption.dropna(subset = ['证券代码'])
    try:
        multiple = redumption[redumption['证券代码'].str.contains(order_book_id[:-5])].iloc[0,-2]/100
    except:
        multiple = 1.3
    
    # Judge whether the time period could be accepted 
    date_pd = pd.Timestamp(date)
    if date_pd > maturity_date:
        #print(bond_name+':time of pricing is after maturity')
        sys.exit(0)
    elif date_pd > de_listed_date:
        #print(bond_name+':time of pricing is after the de-listed date')
        sys.exit(0)
    
    # Get the yeild curve of treasury
    # Get yeild rate of the pricing date (from present to the maturity date)
    yc = rq.get_yield_curve(start_date=date, end_date=date)

    date_pd = pd.Timestamp(date)
    time_remain_month = ((maturity_date - date_pd).days)/30
    time_remain_year = ((maturity_date - date_pd).days)/365

    if time_remain_year <= 1:
        if time_remain_month <= 1:
            risk_free = yc['1M'].iloc[0]
        elif time_remain_month <= 2:
             risk_free = yc['2M'].iloc[0]
        elif time_remain_month <= 3:
            risk_free = yc['3M'].iloc[0]
        elif time_remain_month <= 6:
             risk_free = yc['6M'].iloc[0]
        elif time_remain_month <= 9:
             risk_free = yc['9M'].iloc[0]
        else: 
            risk_free = yc['1Y'].iloc[0]
    else:
        year = math.ceil(time_remain_year)
        risk_free = yc[str(year)+'Y'].iloc[0]
        if str(risk_free) == 'nan':
            risk_free = (yc[str(year-1)+'Y'].iloc[0] + yc[str(year+1)+'Y'].iloc[0] )/2
        
    # Get the stock price
    start = date[:2] + str(int(date[2:4]) - 3) + date[4:]
    stock_price = zw.get_stocks_price([stock_code],start,date).dropna(subset = ['close'])
    spot_price = stock_price['close'].iloc[-1]

    # Calculate volatility(EMA65)
    stock_price['ret'] = stock_price['close']/stock_price['close'].shift(1) - 1
    stock_price['vol'] = pd.Series()
    for i in range(120,len(stock_price)):
        stock_price['vol'].iloc[i] = np.std(stock_price['ret'].iloc[i-120:i])    

    stock_price['vol_ema'] = ta.EMA(stock_price['vol'],65)
    spot_vol = stock_price['vol_ema'][-1]*np.sqrt(252)
    
    # Get the conversion price table
    conversion_price = rq.convertible.get_conversion_price(order_book_id)
    
    # Get the conversion price of the pricing date
    date_pd = pd.Timestamp(date)
    strike_price = -1

    if date_pd >= conversion_price.index[-1][1]:
        strike_price = conversion_price['conversion_price'].iloc[-1]
    else:
        for i in range(len(conversion_price) - 1):
            if date_pd >= conversion_price.index[i][1] and date_pd < conversion_price.index[i+1][1]:
                strike_price = conversion_price['conversion_price'].iloc[i]
                break

    if strike_price == -1:
        #print(bond_name + ':计算时间在初次转股价格发布日之前')
        sys.exit(0) 

    
    ### Calculate the value of long option on the stock
    # Modify the format of time according to quantlib
    maturity_date_ql = ql.Date(maturity_date.day, maturity_date.month, maturity_date.year)
    day_count = ql.Actual365Fixed()
    calendar = ql.China()
    calculation_date = ql.Date(date_pd.day, date_pd.month, date_pd.year)
    ql.Settings.instance().evaluationDate = calculation_date
    
    # Build American option
    option_type = ql.Option.Call
    payoff = ql.PlainVanillaPayoff(option_type, strike_price)
    settlement = calculation_date

    am_exercise = ql.AmericanExercise(settlement, maturity_date_ql)
    american_option = ql.VanillaOption(payoff, am_exercise)
    
    # Black-Scholes-Merton process
    spot_handle = ql.QuoteHandle(ql.SimpleQuote(spot_price))
    flat_ts = ql.YieldTermStructureHandle(ql.FlatForward(calculation_date, risk_free, day_count))
    dividend_yield = ql.YieldTermStructureHandle(ql.FlatForward(calculation_date, dividend_rate, day_count))
    flat_vol_ts = ql.BlackVolTermStructureHandle(ql.BlackConstantVol(calculation_date, calendar, spot_vol, day_count))
    bsm_process = ql.BlackScholesMertonProcess(spot_handle, 
                                               dividend_yield, 
                                               flat_ts, 
                                               flat_vol_ts)    
    
    # Calculate option price using CRR approach
    steps = 400
    binomial_engine = ql.BinomialVanillaEngine(bsm_process, "crr", steps)
    american_option.setPricingEngine(binomial_engine)
    
    # Convert the option price to per 100 convertible bond
    option_position1 = round((100/spot_price) * american_option.NPV(),2)
    delta_position1 =  round((100/spot_price) * american_option.delta(),2)
    gamma_position1 = round((100/spot_price) * american_option.gamma(),2)
    
            
    # calculate vega
    # calculate p_u
    flat_vol_ts = ql.BlackVolTermStructureHandle(ql.BlackConstantVol(calculation_date, calendar, spot_vol+ 0.0001, day_count))
    bsm_process = ql.BlackScholesMertonProcess(spot_handle, 
                                               dividend_yield, 
                                               flat_ts, 
                                               flat_vol_ts)
    steps = 400
    binomial_engine = ql.BinomialVanillaEngine(bsm_process, "crr", steps)
    american_option.setPricingEngine(binomial_engine)
    p_u = american_option.NPV()

    # calculate p_d
    flat_vol_ts = ql.BlackVolTermStructureHandle(ql.BlackConstantVol(calculation_date, calendar, spot_vol- 0.0001, day_count))
    bsm_process = ql.BlackScholesMertonProcess(spot_handle, 
                                               dividend_yield, 
                                               flat_ts, 
                                               flat_vol_ts)
    steps = 400
    binomial_engine = ql.BinomialVanillaEngine(bsm_process, "crr", steps)
    american_option.setPricingEngine(binomial_engine)
    p_d = american_option.NPV()
    vega = (p_u - p_d)/(2*0.0001)
    
    vega_position1 = round((100/spot_price) * vega,2)
    
    
    
    # calculate the short value of long option when strike cap is 130%
    option_type = ql.Option.Call
    payoff = ql.PlainVanillaPayoff(option_type, strike_price*multiple)
    settlement = calculation_date

    am_exercise = ql.AmericanExercise(settlement, maturity_date_ql)
    american_option = ql.VanillaOption(payoff, am_exercise)
    
    spot_handle = ql.QuoteHandle(ql.SimpleQuote(spot_price))
    flat_ts = ql.YieldTermStructureHandle(ql.FlatForward(calculation_date, risk_free, day_count))
    dividend_yield = ql.YieldTermStructureHandle(ql.FlatForward(calculation_date, dividend_rate, day_count))
    flat_vol_ts = ql.BlackVolTermStructureHandle(ql.BlackConstantVol(calculation_date, calendar, spot_vol, day_count))
    bsm_process = ql.BlackScholesMertonProcess(spot_handle, 
                                               dividend_yield, 
                                               flat_ts, 
                                               flat_vol_ts)    
    
    # Calculate price using CRR approach.
    steps = 400
    binomial_engine = ql.BinomialVanillaEngine(bsm_process, "crr", steps)
    american_option.setPricingEngine(binomial_engine)
    option_position2 = round((100/spot_price) * american_option.NPV(),2)
    delta_position2 =  round((100/spot_price) * american_option.delta(),2)
    gamma_position2 = round((100/spot_price) * american_option.gamma(),2)
    
    # calculate vega
    # Calculate p_u
    flat_vol_ts = ql.BlackVolTermStructureHandle(ql.BlackConstantVol(calculation_date, calendar, spot_vol+ 0.0001, day_count))
    bsm_process = ql.BlackScholesMertonProcess(spot_handle, 
                                               dividend_yield, 
                                               flat_ts, 
                                               flat_vol_ts)
    steps = 400
    binomial_engine = ql.BinomialVanillaEngine(bsm_process, "crr", steps)
    american_option.setPricingEngine(binomial_engine)
    p_u = american_option.NPV()

    # Calculate p_d
    flat_vol_ts = ql.BlackVolTermStructureHandle(ql.BlackConstantVol(calculation_date, calendar, spot_vol- 0.0001, day_count))
    bsm_process = ql.BlackScholesMertonProcess(spot_handle, 
                                               dividend_yield, 
                                               flat_ts, 
                                               flat_vol_ts)
    steps = 400
    binomial_engine = ql.BinomialVanillaEngine(bsm_process, "crr", steps)
    american_option.setPricingEngine(binomial_engine)
    p_d = american_option.NPV()
    vega = (p_u - p_d)/(2*0.0001)
    
    vega_position2 = round((100/spot_price) * vega,2)    

    
    # If the pricing date is pre exercise date, we should offset option value at this period
    if date_pd < conversion_start_date:
        # get the risk free rate
        date_pd = pd.Timestamp(date)
        time_remain_month = ((conversion_start_date - date_pd).days)/30
        time_remain_year = ((conversion_start_date - date_pd).days)/365

        if time_remain_year <= 1:
            if time_remain_month <= 1:
                risk_free = yc['1M'].iloc[0]
            elif time_remain_month <= 2:
                 risk_free = yc['2M'].iloc[0]
            elif time_remain_month <= 3:
                risk_free = yc['3M'].iloc[0]
            elif time_remain_month <= 6:
                 risk_free = yc['6M'].iloc[0]
            elif time_remain_month <= 9:
                 risk_free = yc['9M'].iloc[0]
            else: 
                risk_free = yc['1Y'].iloc[0]
        else:
            year = math.ceil(time_remain_year)
            risk_free = yc[str(year)+'Y'].iloc[0]
            if str(risk_free) == 'nan':
                risk_free = (yc[str(year-1)+'Y'].iloc[0] + yc[str(year+1)+'Y'].iloc[0] )/2
    
        # Modify the format of time according to quantlib
        maturity_date_ql = ql.Date(conversion_start_date.day, conversion_start_date.month, conversion_start_date.year)
        day_count = ql.Actual365Fixed()
        calendar = ql.China()
        calculation_date = ql.Date(date_pd.day, date_pd.month, date_pd.year)
        ql.Settings.instance().evaluationDate = calculation_date

        # Build American option
        option_type = ql.Option.Call
        payoff = ql.PlainVanillaPayoff(option_type, strike_price)
        settlement = calculation_date

        am_exercise = ql.AmericanExercise(settlement, maturity_date_ql)
        american_option = ql.VanillaOption(payoff, am_exercise)

        #  Black-Scholes-Merton process
        spot_handle = ql.QuoteHandle(ql.SimpleQuote(spot_price))
        flat_ts = ql.YieldTermStructureHandle(ql.FlatForward(calculation_date, risk_free, day_count))
        dividend_yield = ql.YieldTermStructureHandle(ql.FlatForward(calculation_date, dividend_rate, day_count))
        flat_vol_ts = ql.BlackVolTermStructureHandle(ql.BlackConstantVol(calculation_date, calendar, spot_vol, day_count))
        bsm_process = ql.BlackScholesMertonProcess(spot_handle, 
                                                   dividend_yield, 
                                                   flat_ts, 
                                                   flat_vol_ts)    

        # Calculate price using CRR approach.
        steps = 400
        binomial_engine = ql.BinomialVanillaEngine(bsm_process, "crr", steps)
        american_option.setPricingEngine(binomial_engine)

        # Convert the option price to per 100 convertible bond
        option_position3 = round((100/spot_price) * american_option.NPV(),2)
        delta_position3 =  round((100/spot_price) * american_option.delta(),2)
        gamma_position3 = round((100/spot_price) * american_option.gamma(),2)


        # calculate vega
        # calculate p_u
        flat_vol_ts = ql.BlackVolTermStructureHandle(ql.BlackConstantVol(calculation_date, calendar, spot_vol+ 0.0001, day_count))
        bsm_process = ql.BlackScholesMertonProcess(spot_handle, 
                                                   dividend_yield, 
                                                   flat_ts, 
                                                   flat_vol_ts)
        steps = 400
        binomial_engine = ql.BinomialVanillaEngine(bsm_process, "crr", steps)
        american_option.setPricingEngine(binomial_engine)
        p_u = american_option.NPV()

        # calculate p_d
        flat_vol_ts = ql.BlackVolTermStructureHandle(ql.BlackConstantVol(calculation_date, calendar, spot_vol- 0.0001, day_count))
        bsm_process = ql.BlackScholesMertonProcess(spot_handle, 
                                                   dividend_yield, 
                                                   flat_ts, 
                                                   flat_vol_ts)
        steps = 400
        binomial_engine = ql.BinomialVanillaEngine(bsm_process, "crr", steps)
        american_option.setPricingEngine(binomial_engine)
        p_d = american_option.NPV()
        vega = (p_u - p_d)/(2*0.0001)

        vega_position3 = round((100/spot_price) * vega,2)
        
        is_pre_exercise ='yes'
        
        
        # Build American option
        option_type = ql.Option.Call
        payoff = ql.PlainVanillaPayoff(option_type, strike_price*multiple)
        settlement = calculation_date

        am_exercise = ql.AmericanExercise(settlement, maturity_date_ql)
        american_option = ql.VanillaOption(payoff, am_exercise)

        #  Black-Scholes-Merton process
        spot_handle = ql.QuoteHandle(ql.SimpleQuote(spot_price))
        flat_ts = ql.YieldTermStructureHandle(ql.FlatForward(calculation_date, risk_free, day_count))
        dividend_yield = ql.YieldTermStructureHandle(ql.FlatForward(calculation_date, dividend_rate, day_count))
        flat_vol_ts = ql.BlackVolTermStructureHandle(ql.BlackConstantVol(calculation_date, calendar, spot_vol, day_count))
        bsm_process = ql.BlackScholesMertonProcess(spot_handle, 
                                                   dividend_yield, 
                                                   flat_ts, 
                                                   flat_vol_ts)    

        # Calculate price using CRR approach.
        steps = 400
        binomial_engine = ql.BinomialVanillaEngine(bsm_process, "crr", steps)
        american_option.setPricingEngine(binomial_engine)

        # Convert the option price to per 100 convertible bond
        option_position4 = round((100/spot_price) * american_option.NPV(),2)
        delta_position4 =  round((100/spot_price) * american_option.delta(),2)
        gamma_position4 = round((100/spot_price) * american_option.gamma(),2)


        # calculate vega
        # calculate p_u
        flat_vol_ts = ql.BlackVolTermStructureHandle(ql.BlackConstantVol(calculation_date, calendar, spot_vol+ 0.0001, day_count))
        bsm_process = ql.BlackScholesMertonProcess(spot_handle, 
                                                   dividend_yield, 
                                                   flat_ts, 
                                                   flat_vol_ts)
        steps = 400
        binomial_engine = ql.BinomialVanillaEngine(bsm_process, "crr", steps)
        american_option.setPricingEngine(binomial_engine)
        p_u = american_option.NPV()

        # calculate p_d
        flat_vol_ts = ql.BlackVolTermStructureHandle(ql.BlackConstantVol(calculation_date, calendar, spot_vol- 0.0001, day_count))
        bsm_process = ql.BlackScholesMertonProcess(spot_handle, 
                                                   dividend_yield, 
                                                   flat_ts, 
                                                   flat_vol_ts)
        steps = 400
        binomial_engine = ql.BinomialVanillaEngine(bsm_process, "crr", steps)
        american_option.setPricingEngine(binomial_engine)
        p_d = american_option.NPV()
        vega = (p_u - p_d)/(2*0.0001)

        vega_position4 = round((100/spot_price) * vega,2)                
        
    else:
        option_position3 = 0
        delta_position3 =  0
        gamma_position3 = 0
        vega_position3 = 0
        option_position4 = 0
        delta_position4 =  0
        gamma_position4 = 0
        vega_position4 = 0
        is_pre_exercise ='no'
    
    option_position = option_position1 - option_position2 - option_position3 + option_position4
    delta_position = delta_position1 - delta_position2 - delta_position3 + delta_position4
    gamma_position = gamma_position1 - gamma_position2 - gamma_position3 + gamma_position4
    vega_position = vega_position1 - vega_position2 - vega_position3 + vega_position4
    
    return option_position1,option_position2, option_position3,option_position4,delta_position, gamma_position, vega_position,is_pre_exercise

def get_bond_price(bond_name, date):

    
    # get the id of convertible bonds
    all_info = rq.convertible.all_instruments()
    order_book_id = all_info[all_info['symbol'] == bond_name]['order_book_id'].iloc[0]
    maturity_date =  all_info[all_info['symbol'] == bond_name]['maturity_date'].iloc[0]
    
    # gte the yeild curve of pricing date
    yc = rq.get_yield_curve(start_date=date, end_date=date)
    
    # get the risk-free rate of pricing date
    date_pd = pd.Timestamp(date)
    time_remain_month = ((maturity_date - date_pd).days)/30
    time_remain_year = ((maturity_date - date_pd).days)/365

    if time_remain_year <= 1:
        if time_remain_month <= 1:
            risk_free = yc['1M'].iloc[0]
        elif time_remain_month <= 2:
             risk_free = yc['2M'].iloc[0]
        elif time_remain_month <= 3:
            risk_free = yc['3M'].iloc[0]
        elif time_remain_month <= 6:
             risk_free = yc['6M'].iloc[0]
        elif time_remain_month <= 9:
             risk_free = yc['9M'].iloc[0]
        else: 
            risk_free = yc['1Y'].iloc[0]
    else:
        year = math.ceil(time_remain_year)
        risk_free = yc[str(year)+'Y'].iloc[0]
        if str(risk_free) == 'nan':
            risk_free = (yc[str(year-1)+'Y'].iloc[0] + yc[str(year+1)+'Y'].iloc[0] )/2    
    
    # get the cash flow table of convertible bonds
    cf = convertible.get_cash_flow(order_book_id)
    
    # find the nearest cash flow
    first_cf = -1

    if date_pd <= cf.index[0][1]:
        first_cf = 0
    else:
        for i in range(len(cf.index) - 1):
            if date_pd > cf.index[i][1] and date_pd <= cf.index[i+1][1]:
                first_cf = i + 1
                break

    if first_cf == -1:
        #print(bond_name + ':输入日期超出债券退市日期')
        sys.exit(0)
    
    # calculate bond value
    bond_value = 0
    for i in range(first_cf, len(cf.index)):
        time_diff = (cf.index[i][1] - date_pd).days
        discount = cf['cash_flow_pretax'].iloc[i]/np.power(1+risk_free, time_diff/365)
        bond_value = bond_value + discount
    
    return bond_value

def get_convertible_bond_price(bond_name, date):
    # get the id of convertible bonds
    all_info = rq.convertible.all_instruments()
    order_book_id = all_info[all_info['symbol'] == bond_name]['order_book_id'].iloc[0]    
    bond_price = rq.get_price(order_book_id, start_date=date, end_date=date)['close'][0]
    
    return bond_price

def chunks(List, n):
    return [List[i:i + n] for i in range(0, len(List), n)]

def get_table(trading_date,all_info):
    for date in trading_date:
        select = all_info[all_info['maturity_date'] > pd.Timestamp(date)]
        bond_list = select['symbol']

        stats = []
        for bond in bond_list:
            try:
                maturity_date = str(all_info[all_info['symbol'] == bond]['maturity_date'].iloc[0])[:10]
                bond_price = get_bond_price(bond, date)
                option_position1,option_position2, option_position3, option_position4, delta_position, gamma_position, vega_position,is_pre_exercise = get_convertible_call_price(bond, date)
                convertible_price = bond_price + option_position1 - option_position2 - option_position3 +  option_position4
                real_price = get_convertible_bond_price(bond,  date)
                stats.append([bond,date,maturity_date,is_pre_exercise,bond_price, 
                              option_position1,option_position2,option_position3,option_position4, 
                              convertible_price, real_price, delta_position, gamma_position, vega_position])
            except:
                continue
        #df_stats = pd.DataFrame(stats, columns = ['bond_name','calculation_date','maturity_date','is_pre_exercise','bond_price','option1','option2','option3','option4',
        #                                      'calculation_price','real_price'])
        print(date + ' finished')
        return stats
```

# The relationship between binomial-tree price and market price at the first day of listing

To evaluate whether the model price can accurately reflect the real price, we first examine their relationship on the first day of listing.

We use cal_to_real (model price to real price) to show this relationship.

From the results, the average cal_to_real ratio on the first day of listing is 0.891088. The time theory of the ratio is stationary (verified by ADF test).

This result indicates that on the first day of listing, our binomial-tree model systematically undervalues the price of convertible bonds.

The potential cause could be:
+ (1)	We ignore the reset clause and put clause of convertible bonds, leading to some imperfection of our model
+ (2)	The assumption regards risk-free rate and volatility may slightly deviate from those of the real market.
+ (3)	Investors are over-confident on the value of convertible bonds in general.

The calculation process is shown below:


```python
# Testing the relathionship on the first day of listing
all_info = rq.convertible.all_instruments()
bond_list = all_df['bond_name'].drop_duplicates()
first_open_list =[]
for bond in bond_list:
    bond_info = all_info[all_info['symbol'] == bond]
    value_date = str(bond_info['value_date'].iloc[0])[:10]
    order_book_id = bond_info['order_book_id'].iloc[0]
    # get the information on the first day of listing
    first_day = rq.get_price(order_book_id, start_date=value_date, end_date='2019-04-01', frequency='1d', fields=None, 
                             adjust_type='none', skip_suspended =False, market='cn')
    first_date = str(first_day.index[0])[:10]
    first_open = first_day['open'].iloc[0]
    # get the calculation price on the first day of listing
    try:
        first_pricing = all_df[(all_df['bond_name'] == bond) & (all_df['calculation_date'] == first_date)]['calculation_price'].iloc[0]
        first_open_list.append([bond,first_date,first_open,first_pricing])
    except:
        maturity_date = str(all_info[all_info['symbol'] == bond]['maturity_date'].iloc[0])[:10]
        bond_price = get_bond_price(bond, first_date)
        option_position1,option_position2, option_position3, option_position4, delta_position, gamma_position, vega_position,is_pre_exercise = get_convertible_call_price(bond, first_date)
        convertible_price = bond_price + option_position1 - option_position2 - option_position3 +  option_position4
        first_open_list.append([bond,first_date,first_open,convertible_price])
```

```python
# stationary test using ADF test under 95% confidence level
from statsmodels.tsa.stattools import adfuller
def check_for_stationarity(X, cutoff=0.05):
    # We must observe significant p-value to convince ourselves that the series is stationary
    pvalue = adfuller(X)[1]
    if pvalue < cutoff:
        #print('p-value = ' + str(pvalue) + '. This time_throry'  +' is likely stationary.')
        return True
    else:
        #print('p-value = ' + str(pvalue) + '. This time_throry' + ' is likely non-stationary.')
        return False
```

Firstly, we select all convertible bonds that went into public between 2017 and April 2019. Then we calculate the average cal_to_real ratio based on the number of trading days after the first listed date, thus inducing the overall trend of all cal_to_real ratio.

In the chart below, [day,number,mean,std,score] separately refer to the number of trading days after listed, the number of convertible bonds included, the mean of all cal_to_real, the stand deviation of cal_to_real and score = mean + 1.645std


```python
# the average trend of all cal_to_real ratio during the duration of bonds
bond_list_2 = df_first_open['bond_name'][24:]
#ratio_sum = pd.Series( [0 for x in range(len(zw.get_tradingDates('2017-01-01','2019-04-01')))])
ratio_sum = pd.DataFrame(index =  range(len(zw.get_tradingDates('2017-01-01','2019-04-01'))))
for bond in bond_list_2:
    ratio_ts = all_df[all_df['bond_name'] == bond].reset_index(drop = True)
    ratio_ts = ratio_ts['cal_to_real']
    ratio_sum[bond] = pd.Series(ratio_ts, index = ratio_sum.index)
res = []
for i in range(len(ratio_sum)):
    select = ratio_sum.iloc[i].dropna()
    number = len(select)
    mean = select.mean()
    std = select.std()
    res.append([i,number,mean,std])
df_res = pd.DataFrame(res, columns = ['day','number','mean','std'])
df_res['score'] = df_res['mean'] + 1.645*df_res['std']
df_res
```

## Build trading strategy based on cal_to_real ratio

Another way to test whether the modeled price could accurately reflect market price is to build a trading strategy based on it. If the binomial-tree price is a good indicator of market price, when the deviation between the two is high, i.e., the cal_to_real ratio is high, we would expect the market price will repair itself to the modeled price. Thus, when cal_to_real ratio is very high, the market price is undervalued and we could long the convertible bonds to profit. 

Here we use score to evaluate the degree of deviation between the two prices. 

score = mean of all bonds + 1.645std of all bonds(mean + 1.645std is the upper boundary of 90% confidence interval under normal distribution).

For each date, if the cal_to_real ratio of bond A is higher than the score of that day, we would long the bonds A. 


```python
plt.figure(figsize=[12.,6])
for bond in ratio_sum_30.columns:
    plt.plot(ratio_sum_30.index,ratio_sum_30[bond],linewidth=1)
plt.title('所有转债存续期内cal_to_real变动情况')
plt.show()
```

For the chart above:
    
horizontal axis -- number of days after first listed date

longitudinal axis -- cal_to_real ratio

Each line represents a series of cal_to_real ratio for an individual bonds
    


The code below selects the bonds that has cal_to_real ratio higher than score in some time.

Only 20 of 150 bonds satisfy our criterion. The occurrence of these peak cal_to_real ratio is continuous for days, suggesting that when the extreme happens, the market price doesn’t revert immediately.  

```python
std_upper = []
for i in range(len(ratio_sum_30)):
    for bond in ratio_sum_30.columns:
        if ratio_sum_30[bond].iloc[i] > df_res['score'].iloc[i]:
               std_upper.append([i,bond])
df_std = pd.DataFrame(std_upper,columns=['days', 'bond'])
df_std
```

Calculating the annualized return of the 20 bonds by holding long position when the extreme cal_to_real ratio emerges and closing the position when the extreme ratio disappears.

The results are shown below.

```python
keep_first = df_std.drop_duplicates(subset = 'bond',keep = 'first')
keep_last = df_std.drop_duplicates(subset = 'bond',keep = 'last')
max_ret = []
for bond in df_std['bond'].drop_duplicates():
    first_max = keep_first[keep_first['bond'] == bond]['days'].iloc[0]
    last_max = keep_last[keep_last['bond'] == bond]['days'].iloc[0]
    ret_max = (all_df[all_df['bond_name'] == bond]['real_price'].iloc[last_max] / 
               all_df[all_df['bond_name'] == bond]['real_price'].iloc[first_max]) - 1
    ret_max_annual = np.power(1+ret_max,252/(last_max - first_max)) - 1
    max_ret.append([bond, last_max - first_max,  ret_max_annual])
df_max_ret = pd.DataFrame(max_ret, columns= ['bond','max_days','annual_ret'])
df_max_ret.sort_values(by = ['annual_ret'], ascending= False)
```

```python
max_mean = df_max_ret['annual_ret'].mean()
win_percent = len(df_max_ret[df_max_ret['annual_ret']>0]) / len(df_max_ret)
print('从出现极值到极值消失，所有转债的平均年化收益率为'+str(max_mean)+', 胜率为'+str(win_percent))
```

For this strategy, the annualized return of all bonds is 9% and the chance of profiting is 57%. As shown above, we could profit based on the extreme number of cal_to_real ratio, but profitability is not stable for different bonds and time, suggesting that our model to measure the market price is not highly accurate.
